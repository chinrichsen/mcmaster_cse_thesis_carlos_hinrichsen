```python
import random
import numpy as np
import copy
import pandas as pd
import time
import itertools
from sklearn.cluster import KMeans
from itertools import product
import math


class VRP:
    def __init__(self, num_vehicles, num_customers, locations, demands, vehicle_capacity, rcl_factor=0.98, iterations=10000, local_search_option='2-opt', stagnation_limit=50, perturb_percentage=0.1, split_penalty=1000, sdd = 1, reintegrate_it=1000, split_sol = 'Yes', min_prob_matrix = 1.0, max_prob_matrix = 1.0, int_prob = None, threshold_increment = 5, optimization_solution_finder = "tier"): #, dem_increment = 0.25):
        #random.seed(sdd)
        random.seed()
        #np.random.seed(sdd)
        np.random.seed()
        self.depot_id = 0
        self.number_vehicle = num_vehicles
        self.number_customers = num_customers
        self.number_nodes = num_customers + 1
        self.coordinates = locations
        self.demands = [0] + demands
        self.vehicle_capacity = vehicle_capacity
        self.reintegrate_it = reintegrate_it
        self.saved_initial_solution_split = None
        self.saved_initial_solution_no_split = None
        self.split_sol = split_sol
        self.int_prob = int_prob
        self.optimization_solution_finder = optimization_solution_finder
        #self.dem_increment = dem_increment

        # Set hyperparameters
        self.rcl_factor = rcl_factor
        self.iterations = iterations
        self.local_search_option = local_search_option
        self.stagnation_limit = stagnation_limit
        self.perturb_percentage = perturb_percentage
        self.split_penalty = split_penalty
        self.sdd = sdd
        self.min_prob_matrix = min_prob_matrix
        self.max_prob_matrix = max_prob_matrix
        self.threshold_increment = threshold_increment

        self.coordinates_np = np.array(self.coordinates)
        self.demands = np.array(self.demands)
        diff = self.coordinates_np[:, np.newaxis, :] - self.coordinates_np
        self.distance_matrix = np.sqrt((diff**2).sum(axis=2))

    def cluster_based_init(self, n_clusters=None):
        if n_clusters is None:
            n_clusters = self.number_vehicle  # Or use a heuristic to determine the number of clusters
        # Apply clustering on customer locations only
        customer_coords = np.array(self.coordinates[1:])  # Exclude depot
        kmeans = KMeans(n_clusters=n_clusters, random_state=self.sdd).fit(customer_coords)
        clusters = kmeans.labels_

        remaining_demands = self.demands.copy()
        routes = []

        for cluster_id in range(n_clusters):
            cluster_customers = [i+1 for i, cluster in enumerate(clusters) if cluster == cluster_id and remaining_demands[i+1] > 0]
            # Apply nearest neighbor within each cluster
            cluster_routes, cluster_remaining_demands = self.nearest_neighbor_within_cluster(cluster_customers, remaining_demands)
            routes.extend(cluster_routes)
            remaining_demands = cluster_remaining_demands

        return routes, remaining_demands

    def nearest_neighbor_within_cluster(self, cluster_customers, remaining_demands):
        cluster_routes = []
        demands_copy = remaining_demands.copy()

        while any(demands_copy[customer] > 0 for customer in cluster_customers):
            route = [self.depot_id]
            current_load = 0
            current_location = self.depot_id

            while current_load < self.vehicle_capacity:
                nearest_customer = None
                nearest_distance = float('inf')

                for customer in cluster_customers:
                    if demands_copy[customer] > 0 and customer not in route:
                        distance = self.distance_matrix[current_location][customer]
                        if distance < nearest_distance and current_load + demands_copy[customer] <= self.vehicle_capacity:
                            nearest_distance = distance
                            nearest_customer = customer

                if nearest_customer is None:
                    break  # No more customers can be added without exceeding capacity

                current_load += demands_copy[nearest_customer]
                demands_copy[nearest_customer] = 0
                route.append(nearest_customer)
                current_location = nearest_customer

            route.append(self.depot_id)
            cluster_routes.append(route)

        return cluster_routes, demands_copy


    def nearest_neighbor(self):
        remaining_demands = self.demands.copy()
        routes = []
        fes_cond = None

        for _ in range(self.number_vehicle):
            current_load = 0
            current_location = self.depot_id
            route = [self.depot_id]

            while current_load < self.vehicle_capacity:
                nearest_customer = None
                nearest_distance = float('inf')

                for i in range(1, self.number_customers + 1):
                    if remaining_demands[i] > 0 and i not in route:
                        distance = self.distance_matrix[current_location][i]
                        if distance < nearest_distance and current_load + remaining_demands[i] <= self.vehicle_capacity:
                            nearest_distance = distance
                            nearest_customer = i

                if nearest_customer is None:
                    break

                current_load += remaining_demands[nearest_customer]
                remaining_demands[nearest_customer] = 0
                route.append(nearest_customer)
                current_location = nearest_customer

            route.append(self.depot_id)
            routes.append(route)
            serviced_demands = []
            for route in routes:
                route_demands = {}
                for customer in route[1:-1]:  # Exclude the depot
                    route_demands[customer] = self.demands[customer]
                serviced_demands.append(route_demands)
        # Check if initial solution is feasible
        # Flatten the list
        flat_list = [item for sublist in routes for item in sublist]

        # Convert the flat list to a set to get unique elements
        unique_elements = set(flat_list)

        # Count the number of unique elements
        count_unique = len(unique_elements)
        if num_customers+1 != count_unique:
            fes_cond = 'Infeasible'
            print("Infeasible construction solution")
        else:
            fes_cond = 'Feasible'
            print("Feasible construction solution")
        if self.saved_initial_solution_no_split is None:# and fes_cond == 'Feasible':
            # Save the initial solution right after it's generated, ensuring no split deliveries
            self.saved_initial_solution_no_split = (copy.deepcopy(routes), copy.deepcopy(remaining_demands), copy.deepcopy(serviced_demands), copy.deepcopy(fes_cond))

        return routes, remaining_demands, serviced_demands, fes_cond

    def nearest_neighbor_with_split_demands(self):
        remaining_demands = self.demands.copy()
        routes = []
        serviced_demands = []
        fes_cond = None

        for _ in range(self.number_vehicle):
            current_load = 0
            current_location = self.depot_id
            route = [self.depot_id]
            route_demands = {}

            while current_load < self.vehicle_capacity:
                nearest_customer = None
                nearest_distance = float('inf')

                for i in range(1, self.number_customers + 1):
                    if remaining_demands[i] > 0 and i not in route:
                        distance = self.distance_matrix[current_location][i]
                        if distance < nearest_distance:
                            nearest_distance = distance
                            nearest_customer = i

                if nearest_customer is None:
                    break

                # Determine how much of the demand can be fulfilled by this vehicle
                demand_to_fulfill = min(remaining_demands[nearest_customer], self.vehicle_capacity - current_load)
                if demand_to_fulfill > 0:  # If there's some demand that can be fulfilled
                    current_load += demand_to_fulfill
                    remaining_demands[nearest_customer] -= demand_to_fulfill
                    route_demands[nearest_customer] = demand_to_fulfill

                    if nearest_customer not in route:  # Add to route if not already added
                        route.append(nearest_customer)
                        current_location = nearest_customer

            route.append(self.depot_id)
            routes.append(route)
            serviced_demands.append(route_demands)

        # Prepare the serviced demands in the desired format
        combined_serviced_demands = []
        for route_demands in serviced_demands:
            combined_serviced_demands.append(route_demands)

        # Flatten the list
        flat_list = [item for sublist in routes for item in sublist]

        # Convert the flat list to a set to get unique elements
        unique_elements = set(flat_list)

        # Count the number of unique elements
        count_unique = len(unique_elements)
        if num_customers+1 != count_unique:
            fes_cond = 'Infeasible'
        else:
            fes_cond = 'Feasible'
        
        if self.saved_initial_solution_split is None:# and fes_cond == 'Feasible':
            # Save the initial solution right after it's generated, allowing for split deliveries
            self.saved_initial_solution_split = (copy.deepcopy(routes), copy.deepcopy(remaining_demands), copy.deepcopy(combined_serviced_demands), copy.deepcopy(fes_cond))

        return routes, remaining_demands, combined_serviced_demands, fes_cond    


    def farthest_customer_init(self):
        remaining_demands = self.demands.copy()
        routes = []

        while np.any(remaining_demands[1:] > 0):  # Check if there's any customer demand left
            route = [self.depot_id]
            current_load = 0
            distances_to_depot = self.distance_matrix[self.depot_id][1:]  # Excludes depot itself

            # Prioritize customers based on distance, but consider only those that can be added without exceeding capacity
            eligible_customers = [(i+1, dist) for i, dist in enumerate(distances_to_depot) if remaining_demands[i+1] > 0 and remaining_demands[i+1] <= self.vehicle_capacity - current_load]
            if not eligible_customers:
                break  # If no eligible customers, break

            # Sort eligible customers by distance in descending order (farthest first)
            farthest_customer = sorted(eligible_customers, key=lambda x: x[1], reverse=True)[0][0]

            # Add farthest customer as the first customer of the route
            current_load += remaining_demands[farthest_customer]
            remaining_demands[farthest_customer] = 0
            route.append(farthest_customer)

            while current_load < self.vehicle_capacity:
                nearest_customer = None
                nearest_distance = float('inf')

                for i in range(1, self.number_customers + 1):
                    if remaining_demands[i] > 0 and i not in route:
                        distance = self.distance_matrix[route[-1]][i]  # Distance from the last customer added to potential next customer
                        if distance < nearest_distance and current_load + remaining_demands[i] <= self.vehicle_capacity:
                            nearest_distance = distance
                            nearest_customer = i

                if nearest_customer is None:
                    break  # No more customers can be added without exceeding capacity

                current_load += remaining_demands[nearest_customer]
                remaining_demands[nearest_customer] = 0
                route.append(nearest_customer)

            route.append(self.depot_id)
            routes.append(route)

        return routes, remaining_demands


    def count_split_deliveries(self, routes, route_demands):
        """ Count the number of split deliveries for the given routes and demands. """
        split_count = 0
        deliveries = {}

        for i, demands_dict in enumerate(route_demands):
            for customer, demand in demands_dict.items():
                if customer not in deliveries:
                    deliveries[customer] = 0
                deliveries[customer] += 1

        for customer, delivery_count in deliveries.items():
            if delivery_count > 1:
                split_count += delivery_count - 1  # We add (delivery_count - 1) because if a customer is served by 2 routes, then there is 1 split delivery for that customer.

        return split_count

    def greedy_cost(self, customer_id, current_load):
        return self.demands[customer_id] / (self.distance_matrix[self.depot_id, customer_id] + 1e-5)




    def construct_solution(self, reintegrate=False):
        #print(split_sol)
        # --> name of method: nearest_neighbor_with_split_demands
        if reintegrate:
        # Reintegrate the saved initial solution without modifications
            if split_sol == 'No':
                routes, remaining_demands, serviced_demands, fes_cond = copy.deepcopy(self.saved_initial_solution_no_split)
            else:
                routes, remaining_demands, serviced_demands, fes_cond = copy.deepcopy(self.saved_initial_solution_split)
        else:
            if split_sol == 'No':
                routes, remaining_demands, serviced_demands, fes_cond = self.nearest_neighbor()
                
            else:
                routes, remaining_demands, serviced_demands, fes_cond = self.nearest_neighbor_with_split_demands()

        # Generate a new solution using nearest_neighbor or another method intended for new solutions
            
            #print(routes)
            #print(remaining_demands)
            #print(serviced_demands)

        # First, we use the nearest neighbor method to get the initial routes.
        #routes, remaining_demands, serviced_demands = self.nearest_neighbor()
        #routes, remaining_demands = self.farthest_customer_init()
        #routes, remaining_demands = self.cluster_based_init()
        #print("Routes after cluster solution: ", routes)
        #print("Remaining demands after cluster solution: ", remaining_demands)
        #print("Serviced demands after cluster solution: ", serviced_demands)
        #print(routes)
        #print(remaining_demands)
            serviced_demands = []

            for route in routes:
                route_demands = {}
                for customer in route[1:-1]:  # Exclude the depot
                    route_demands[customer] = self.demands[customer]
                serviced_demands.append(route_demands)

            remaining_demands = self.demands.copy()
            routes = []
            serviced_demands = []

            for _ in range(self.number_vehicle):
                current_load = 0
                route = [self.depot_id]
                route_demands = {}
                while current_load < self.vehicle_capacity:
                    costs = [(i, self.greedy_cost(i, current_load)) for i in range(1, self.number_customers + 1) if remaining_demands[i] > 0 and i not in route]
                    costs.sort(key=lambda x: x[1], reverse=True)
                    rcl_size = max(1, int(len(costs) * self.rcl_factor))
                    rcl = [c[0] for c in costs[:rcl_size]]

                    if not rcl:
                        break

                    random.seed(10)

                    chosen_customer = random.choice(rcl)
                    delivery = min(self.vehicle_capacity - current_load, remaining_demands[chosen_customer])
                    if chosen_customer not in route_demands:
                        route_demands[chosen_customer] = 0
                    route_demands[chosen_customer] += delivery
                    remaining_demands[chosen_customer] -= delivery
                    current_load += delivery
                    route.append(chosen_customer)
                    if current_load == self.vehicle_capacity:
                        break

                route.append(self.depot_id)
                routes.append(route)
                serviced_demands.append(route_demands)

            
        return routes, serviced_demands, remaining_demands

    def is_route_valid(self, route, route_demands):
        total_demand = sum([route_demands[customer] for customer in route if customer in route_demands])
        return total_demand <= self.vehicle_capacity
    
    def filter_demands(self, route_demands, demands):
        # Convert route_demands keys to a set for faster lookup
        route_customers = set(route_demands.keys())
        
        # Filter demands based on route_customers
        filtered_demands = [demand for index, demand in enumerate(demands, start=1) if index in route_customers]
        
        return filtered_demands
    
    # Function to calculate the required values
    def calculate_demands_for_capacity_check(self, demands_list):
        total_demand_by_route = []
        total_demand = 0

        for route_demands in demands_list:
            route_total = sum(route_demands.values())
            total_demand_by_route.append(route_total)
            total_demand += route_total

        number_of_routes = len(demands_list)
        
        return total_demand_by_route, number_of_routes, total_demand

    def swap_local_search(self, routes, route_demands, interdiction_probabilities, iter_num, tier_1_it, tier_2_it):
        improvement_found = True
        it_number = iter_num+1
        
        osf_swl = self.optimization_solution_finder
                                    
        c_it = 1
        while improvement_found:
            # print("Improvement Found in SWAP: ", improvement_found)
            improvement_found = False
            for i in range(len(routes)):
                for j in range(i + 1, len(routes)):
                    for customer_i in routes[i][1:-1]:  # Exclude depot
                        for customer_j in routes[j][1:-1]:  # Exclude depot
                            # Check if customers are not repeated within their new routes before swap
                            c_it = c_it+1
                                        
                            if customer_i not in routes[j] and customer_j not in routes[i]:
                                # Prepare for swap
                                new_route_i = routes[i].copy()
                                new_route_j = routes[j].copy()

                                # Safely get demands or 0 if not found
                                demand_i = route_demands[i].get(customer_i, 0)
                                demand_j = route_demands[j].get(customer_j, 0)

                                # Perform swap
                                new_route_i[new_route_i.index(customer_i)] = customer_j
                                new_route_j[new_route_j.index(customer_j)] = customer_i
                                
                                # Swap demands
                                new_demands_i = route_demands[i].copy()
                                new_demands_j = route_demands[j].copy()
                                new_demands_i[customer_j] = demand_j
                                new_demands_j[customer_i] = demand_i
                                
                                if customer_i in new_demands_i: del new_demands_i[customer_i]
                                if customer_j in new_demands_j: del new_demands_j[customer_j]

                                
                                a1 = self.calculate_cumulative_probabilities_for_routes(routes[i], interdiction_probabilities)
                                a2 = self.calculate_cumulative_probabilities_for_routes(routes[j], interdiction_probabilities)
                                a3 = self.calculate_cumulative_probabilities_for_routes(new_route_i, interdiction_probabilities)
                                a4 = self.calculate_cumulative_probabilities_for_routes(new_route_j, interdiction_probabilities)
                                
                                route_old_i = routes[i]
                                route_old_j = routes[j]
                                route_new_i = new_route_i
                                route_new_j = new_route_j
                                route_demands_old_i = route_demands[i]
                                route_demands_old_j = route_demands[j]
                                route_demands_new_i = new_demands_i
                                route_demands_new_j = new_demands_j

                                
                                route_old_i = self.check_and_transform_path(route_old_i)
                                route_old_j = self.check_and_transform_path(route_old_j)
                                route_new_i = self.check_and_transform_path(route_new_i)
                                route_new_j = self.check_and_transform_path(route_new_j)

                                route_demands_old_i = self.check_and_transform_demands(route_demands_old_i)
                                route_demands_old_j = self.check_and_transform_demands(route_demands_old_j)
                                route_demands_new_i = self.check_and_transform_demands(route_demands_new_i)
                                route_demands_new_j = self.check_and_transform_demands(route_demands_new_j)

                                route_demands_old_i = self.reorder_demands_by_route(route_old_i, route_demands_old_i)
                                route_demands_old_j = self.reorder_demands_by_route(route_old_j, route_demands_old_j)
                                route_demands_new_i = self.reorder_demands_by_route(route_new_i, route_demands_new_i)
                                route_demands_new_j = self.reorder_demands_by_route(route_new_j, route_demands_new_j)

                                
                            
                                # Subset of data source demands to apply vrpi check for 2-opt
                                source_demands_filtered_route_old_i = self.filter_demands(route_demands_old_i[0], demands)
                                source_demands_filtered_route_old_j = self.filter_demands(route_demands_old_j[0], demands)
                                source_demands_filtered_route_new_i = self.filter_demands(route_demands_new_i[0], demands)
                                source_demands_filtered_route_new_j = self.filter_demands(route_demands_new_j[0], demands)

                                
                                revised_demands_old_route_i = self.calculate_revised_demands_for_routes(route_old_i, route_demands_old_i, interdiction_probabilities)
                                revised_demands_old_route_j = self.calculate_revised_demands_for_routes(route_old_j, route_demands_old_j, interdiction_probabilities)
                                revised_demands_new_route_i = self.calculate_revised_demands_for_routes(route_new_i, route_demands_new_i, interdiction_probabilities)
                                revised_demands_new_route_j = self.calculate_revised_demands_for_routes(route_new_j, route_demands_new_j, interdiction_probabilities)
                                
                                all_cumulative_probabilities_old_route_i = self.calculate_cumulative_probabilities_for_routes(route_old_i, interdiction_probabilities)
                                all_cumulative_probabilities_old_route_j = self.calculate_cumulative_probabilities_for_routes(route_old_j, interdiction_probabilities)
                                all_cumulative_probabilities_new_route_i = self.calculate_cumulative_probabilities_for_routes(route_new_i, interdiction_probabilities)
                                all_cumulative_probabilities_new_route_j = self.calculate_cumulative_probabilities_for_routes(route_new_j, interdiction_probabilities)

                                expected_demands_old_route_i = self.calculate_expected_demand_serviced(revised_demands_old_route_i, all_cumulative_probabilities_old_route_i)
                                expected_demands_old_route_j = self.calculate_expected_demand_serviced(revised_demands_old_route_j, all_cumulative_probabilities_old_route_j)
                                expected_demands_new_route_i = self.calculate_expected_demand_serviced(revised_demands_new_route_i, all_cumulative_probabilities_new_route_i)
                                expected_demands_new_route_j = self.calculate_expected_demand_serviced(revised_demands_new_route_j, all_cumulative_probabilities_new_route_j)
                                
                                

                                num_vehicles_old_route_i = sum(1 for item in route_old_i if isinstance(item, list))
                                num_vehicles_old_route_j = sum(1 for item in route_old_j if isinstance(item, list))
                                num_vehicles_new_route_i = sum(1 for item in route_new_i if isinstance(item, list))
                                num_vehicles_new_route_j = sum(1 for item in route_new_j if isinstance(item, list))

                                # Calculate the demands
                                total_demand_by_route_check_old_route_i, number_of_routes_check_old_route_i, total_demand_check_old_route_i = self.calculate_demands_for_capacity_check(revised_demands_old_route_i)
                                
                                total_demand_by_route_check_old_route_j, number_of_routes_check_old_route_j, total_demand_check_old_route_j = self.calculate_demands_for_capacity_check(revised_demands_old_route_j)
                                
                                total_demand_by_route_check_new_route_i, number_of_routes_check_new_route_i, total_demand_check_new_route_i = self.calculate_demands_for_capacity_check(revised_demands_new_route_i)
                                
                                total_demand_by_route_check_new_route_j, number_of_routes_check_new_route_j, total_demand_check_new_route_j = self.calculate_demands_for_capacity_check(revised_demands_new_route_j)
                                
                                # print("Self Route 1 is valid: ", self.is_route_valid(new_route_i, new_demands_i))
                                # print("Self Route 2 is valid: ", self.is_route_valid(new_route_j, new_demands_j))

                                # Check if swap is valid
                                if self.is_route_valid(new_route_i, new_demands_i) and self.is_route_valid(new_route_j, new_demands_j):
                                    cost_before = self.route_cost(routes[i]) + self.route_cost(routes[j])
                                    cost_after = self.route_cost(new_route_i) + self.route_cost(new_route_j)

                                    if osf_swl == "cost":
                                        if cost_after < cost_before:
                                            routes[i], routes[j] = new_route_i, new_route_j
                                            route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                            improvement_found = True

                                    elif osf_swl == "inter_route":
                                        if (a4[0][0]+a3[0][0]<a2[0][0]+a1[0][0]):
                                            routes[i], routes[j] = new_route_i, new_route_j
                                            route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                            improvement_found = True
                                            
                                    elif osf_swl == "min_demand_route":
                                        if ((round(total_demand_check_new_route_i + total_demand_check_new_route_j,1) < round(total_demand_check_old_route_i + total_demand_check_old_route_j,1) )): 
                                            routes[i], routes[j] = new_route_i, new_route_j
                                            route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                            improvement_found = True
                                            
                                    elif osf_swl == "tier":
                                        if it_number <= tier_1_it:
                                            if ((round(total_demand_check_new_route_i + total_demand_check_new_route_j,1) < round(total_demand_check_old_route_i + total_demand_check_old_route_j,1) )): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number > tier_1_it and it_number <= tier_2_it:
                                            if (a4[0][0]+a3[0][0]<a2[0][0]+a1[0][0]): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number > tier_2_it:
                                            if ((cost_after < cost_before)):
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True

                                    elif osf_swl == "random":
                                        a = random.uniform(0, 1)
                                        if a <= 0.33:
                                            if ((round(total_demand_check_new_route_i + total_demand_check_new_route_j,1) < round(total_demand_check_old_route_i + total_demand_check_old_route_j,1) )): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif a > 0.33 and a <= 0.66:
                                            if (a4[0][0]+a3[0][0]<a2[0][0]+a1[0][0]): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif a > 0.66:
                                            if ((cost_after < cost_before)):
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True

                                    elif osf_swl == "mixed":
                                        b = random.uniform(0, 1)
                                        if it_number <= tier_1_it and b <= 0.33:
                                            if ((round(total_demand_check_new_route_i + total_demand_check_new_route_j,1) < round(total_demand_check_old_route_i + total_demand_check_old_route_j,1) )): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number > tier_1_it and it_number <= tier_2_it and b <= 0.33:
                                            if (a4[0][0]+a3[0][0]<a2[0][0]+a1[0][0]): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number > tier_2_it and b <= 0.33:
                                            if ((cost_after < cost_before)):
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number <= tier_1_it and b > 0.33 and b <= 0.66:
                                            if ((round(total_demand_check_new_route_i + total_demand_check_new_route_j,1) < round(total_demand_check_old_route_i + total_demand_check_old_route_j,1) )): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number > tier_1_it and it_number <= tier_2_it and b > 0.33 and b <= 0.66:
                                            if (a4[0][0]+a3[0][0]<a2[0][0]+a1[0][0]): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number > tier_2_it and b > 0.33 and b <= 0.66:
                                            if ((cost_after < cost_before)):
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                        elif it_number <= tier_1_it and b > 0.66:
                                            if ((round(total_demand_check_new_route_i + total_demand_check_new_route_j,1) < round(total_demand_check_old_route_i + total_demand_check_old_route_j,1) )): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                                
                                        elif it_number > tier_1_it and it_number <= tier_2_it and b > 0.66:
                                            
                                            if (a4[0][0]+a3[0][0]<a2[0][0]+a1[0][0]): 
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                                
                                        elif it_number > tier_2_it and b > 0.66:
                                            if ((cost_after < cost_before)):
                                                routes[i], routes[j] = new_route_i, new_route_j
                                                route_demands[i], route_demands[j] = new_demands_i, new_demands_j
                                                improvement_found = True
                                                
                                            
                            



    def two_opt(self, route, route_demands, interdiction_probabilities, iter_num, tier_1_it, tier_2_it):
        best_route = route[:]
        improved = True
        it_number = iter_num+1
        
        osf_swl = self.optimization_solution_finder
        while improved:
            improved = False
            for i in range(1, len(route) - 2):
                for j in range(i + 1, len(route) - 1):
                    c1 = self.calculate_cumulative_probabilities_for_routes(route[:i] + list(reversed(route[i:j+1])) + route[j+1:], interdiction_probabilities)
                    c2 = self.calculate_cumulative_probabilities_for_routes(route[:i] + list((route[i:j+1])) + route[j+1:], interdiction_probabilities)
                    new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                    old_route = route[:i] + list((route[i:j+1])) + route[j+1:]
                    old_demands = route_demands

                    new_route = self.check_and_transform_path(new_route)
                    old_route = self.check_and_transform_path(old_route)
                    old_demands = self.check_and_transform_demands(old_demands)
                    
                    new_demands = self.reorder_demands_by_route(new_route, old_demands)
                    new_demands = self.check_and_transform_demands(new_demands)
                
                    
                    # Subset of data source demands to apply vrpi check for 2-opt
                    source_demands_filtered = self.filter_demands(route_demands, demands)
                    
                    revised_demands_old_route = self.calculate_revised_demands_for_routes(old_route, old_demands, interdiction_probabilities)
                    revised_demands_new_route = self.calculate_revised_demands_for_routes(new_route, new_demands, interdiction_probabilities)
                    all_cumulative_probabilities_old_route = self.calculate_cumulative_probabilities_for_routes(old_route, interdiction_probabilities)
                    all_cumulative_probabilities_new_route = self.calculate_cumulative_probabilities_for_routes(new_route, interdiction_probabilities)
                    
                    expected_demands_old_route = self.calculate_expected_demand_serviced(revised_demands_old_route, all_cumulative_probabilities_old_route)
                    
                    expected_demands_new_route = self.calculate_expected_demand_serviced(revised_demands_new_route, all_cumulative_probabilities_new_route)
                    
                    # Calculate the demands
                    total_demand_by_route_check_old_route, number_of_routes_check_old_route, total_demand_check_old_route = self.calculate_demands_for_capacity_check(revised_demands_old_route)
                    
                    total_demand_by_route_check_new_route, number_of_routes_check_new_route, total_demand_check_new_route = self.calculate_demands_for_capacity_check(revised_demands_new_route)
                    
                    if osf_swl == "tier":
                        if it_number <= tier_1_it:
                            if round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1):
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                        elif it_number > tier_1_it and it_number <= tier_2_it:
                            if c1[0][0] < c2[0][0]:
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True
                        
                        elif it_number > tier_2_it :
                            if (self.distance_matrix[route[i-1], route[i]] + self.distance_matrix[route[j], route[j+1]] > 
                         self.distance_matrix[route[i-1], route[j]] + self.distance_matrix[route[i], route[j+1]]):
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                    elif osf_swl == "cost":
                        if (self.distance_matrix[route[i-1], route[i]] + self.distance_matrix[route[j], route[j+1]] > 
                         self.distance_matrix[route[i-1], route[j]] + self.distance_matrix[route[i], route[j+1]]):
                            new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                            if self.is_route_valid(new_route, route_demands):
                                route = new_route
                                    
                                improved = True

                    elif osf_swl == "inter_route":
                        if c1[0][0] < c2[0][0]:
                            new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                            if self.is_route_valid(new_route, route_demands):
                                route = new_route
                                    
                                improved = True

                    elif osf_swl == "min_demand_route":
                        if round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1):
                            new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                            if self.is_route_valid(new_route, route_demands):
                                route = new_route
                                    
                                improved = True
                    
                    elif osf_swl == "random":
                        a = random.uniform(0, 1)
                        if a <= 0.33:
                            if round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1):
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                        elif a > 0.33 and a <= 0.66:
                            if c1[0][0] < c2[0][0]:
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True
                        
                        elif a > 0.66:
                            if (self.distance_matrix[route[i-1], route[i]] + self.distance_matrix[route[j], route[j+1]] > 
                         self.distance_matrix[route[i-1], route[j]] + self.distance_matrix[route[i], route[j+1]]):
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                    elif osf_swl == "mixed":
                        b = random.uniform(0, 1)
                        if it_number <= tier_1_it and b <= 0.33:
                            if round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1):
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                        elif it_number > tier_1_it and it_number <= tier_2_it and b <= 0.33:
                            if c1[0][0] < c2[0][0]:
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True
                        
                        elif it_number > tier_2_it and b <= 0.33:
                            if (self.distance_matrix[route[i-1], route[i]] + self.distance_matrix[route[j], route[j+1]] > 
                         self.distance_matrix[route[i-1], route[j]] + self.distance_matrix[route[i], route[j+1]]):
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                        elif it_number <= tier_1_it and (b > 0.33 and b <= 0.66):
                            if round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1):
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                        elif it_number > tier_1_it and it_number <= tier_2_it and (b > 0.33 and b <= 0.66):
                            if c1[0][0] < c2[0][0]:
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True
                        
                        elif it_number > tier_2_it and (b > 0.33 and b <= 0.66):
                            if (self.distance_matrix[route[i-1], route[i]] + self.distance_matrix[route[j], route[j+1]] > 
                         self.distance_matrix[route[i-1], route[j]] + self.distance_matrix[route[i], route[j+1]]):
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                        elif it_number <= tier_1_it and b > 0.66:
                            if round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1):
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

                        elif it_number > tier_1_it and it_number <= tier_2_it and b > 0.66:
                            if c1[0][0] < c2[0][0]:
                            
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True
                        
                        elif it_number > tier_2_it and b > 0.66:
                            if (self.distance_matrix[route[i-1], route[i]] + self.distance_matrix[route[j], route[j+1]] > 
                         self.distance_matrix[route[i-1], route[j]] + self.distance_matrix[route[i], route[j+1]]):
                                new_route = route[:i] + list(reversed(route[i:j+1])) + route[j+1:]
                                if self.is_route_valid(new_route, route_demands):
                                    route = new_route
                                        
                                    improved = True

         
        return route

    def three_opt(self, route, route_demands , interdiction_probabilities, iter_num, tier_1_it, tier_2_it):
        best_route = route[:]
        best_distance = self.route_cost(route)
        it_number = iter_num+1
        
        osf_swl = self.optimization_solution_finder
        for i in range(1, len(route) - 2):
            for j in range(i + 1, len(route) - 1):
                for k in range(j + 1, len(route)):
                    new_routes = [
                        route[:i] + route[j:k] + route[i:j] + route[k:],
                        route[:i] + list(reversed(route[i:j])) + route[j:k] + route[k:],
                        route[:i] + route[j:k] + list(reversed(route[i:j])) + route[k:]
                    ]
                    for new_route in new_routes:
                        
                        
                        old_route = best_route
                        old_demands = route_demands

                        new_route_bu = new_route

                        new_route = self.check_and_transform_path(new_route_bu)
                        old_route = self.check_and_transform_path(old_route)
                        old_demands = self.check_and_transform_demands(old_demands)
                        
                        new_demands = self.reorder_demands_by_route(new_route, old_demands)
                        new_demands_bu = new_demands
                        new_demands = self.check_and_transform_demands(new_demands)

                        
                        revised_demands_old_route = self.calculate_revised_demands_for_routes(old_route, old_demands, interdiction_probabilities)
                        revised_demands_new_route = self.calculate_revised_demands_for_routes(new_route, new_demands, interdiction_probabilities)
                        all_cumulative_probabilities_old_route = self.calculate_cumulative_probabilities_for_routes(old_route, interdiction_probabilities)
                        all_cumulative_probabilities_new_route = self.calculate_cumulative_probabilities_for_routes(new_route, interdiction_probabilities)
                        
                        expected_demands_old_route = self.calculate_expected_demand_serviced(revised_demands_old_route, all_cumulative_probabilities_old_route)
                        
                        expected_demands_new_route = self.calculate_expected_demand_serviced(revised_demands_new_route, all_cumulative_probabilities_new_route)
                        c1 = self.calculate_cumulative_probabilities_for_routes(best_route, interdiction_probabilities)
                        c2 = self.calculate_cumulative_probabilities_for_routes(new_route_bu, interdiction_probabilities)
                        

                        total_demand_by_route_check_old_route, number_of_routes_check_old_route, total_demand_check_old_route = self.calculate_demands_for_capacity_check(revised_demands_old_route)
                        
                        total_demand_by_route_check_new_route, number_of_routes_check_new_route, total_demand_check_new_route = self.calculate_demands_for_capacity_check(revised_demands_new_route)
                        
                        new_route = new_route_bu
                        current_distance = self.route_cost(new_route_bu)

                        total_demand_check_new_route < total_demand_check_old_route
                        route_demands = new_demands_bu
                        
                        if osf_swl == "tier":
                            if it_number <= tier_1_it:
                                if ((round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_bu, route_demands)):
                                    
                                    best_route = new_route[:]
                                

                            elif it_number > tier_1_it and it_number <= tier_2_it:
                                if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_bu, route_demands):
                                    best_route = new_route[:]
                                
                            
                            elif it_number > tier_2_it :
                                if current_distance < best_distance and self.is_route_valid(new_route_bu, route_demands):
                                    best_distance = current_distance
                                    best_route = new_route[:]
                               
                        elif osf_swl == "cost":
                            if current_distance < best_distance and self.is_route_valid(new_route_bu, route_demands):
                                best_distance = current_distance
                                best_route = new_route[:]

                        elif osf_swl == "inter_route":
                            if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_bu, route_demands):
                                best_route = new_route[:]

                        elif osf_swl == "min_demand_route":
                            if ((round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_bu, route_demands)):
                                best_route = new_route[:]
                        
                        elif osf_swl == "random":
                            a = random.uniform(0, 1)
                            if a <= 0.33:
                                if ((round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_bu, route_demands)):
                                    best_route = new_route[:]
                                

                            elif a > 0.33 and a <= 0.66:
                                if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_bu, route_demands):
                                    best_route = new_route[:]
                                
                            
                            elif a > 0.66:
                                if current_distance < best_distance and self.is_route_valid(new_route_bu, route_demands):
                                    best_distance = current_distance
                                    best_route = new_route[:]
                                

                        elif osf_swl == "mixed":
                            b = random.uniform(0, 1)
                            if it_number <= tier_1_it and b <= 0.33:
                                if ((round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_bu, route_demands)):
                                    best_route = new_route[:]
                                

                            elif it_number > tier_1_it and it_number <= tier_2_it and b <= 0.33:
                                if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_bu, route_demands):
                                    best_route = new_route[:]
                                
                            
                            elif it_number > tier_2_it and b <= 0.33:
                                if current_distance < best_distance and self.is_route_valid(new_route_bu, route_demands):
                                    best_distance = current_distance
                                    best_route = new_route[:]
                                

                            elif it_number <= tier_1_it and (b > 0.33 and b <= 0.66):
                                if ((round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_bu, route_demands)):
                                    best_route = new_route[:]
                                

                            elif it_number > tier_1_it and it_number <= tier_2_it and (b > 0.33 and b <= 0.66):
                                if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_bu, route_demands):
                                    best_route = new_route[:]
                                
                            
                            elif it_number > tier_2_it and (b > 0.33 and b <= 0.66):
                                if current_distance < best_distance and self.is_route_valid(new_route_bu, route_demands):
                                    best_distance = current_distance
                                    best_route = new_route[:]
                                
                                

                            elif it_number <= tier_1_it and b > 0.66:
                                if ((round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_bu, route_demands)):
                                    best_route = new_route[:]
                                

                            elif it_number > tier_1_it and it_number <= tier_2_it and b > 0.66:
                                if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_bu, route_demands):
                                    best_route = new_route[:]
                                
                            
                            elif it_number > tier_2_it and b > 0.66:
                                if current_distance < best_distance and self.is_route_valid(new_route_bu, route_demands):
                                    best_distance = current_distance
                                    best_route = new_route[:]
                                

        return best_route

    def segment_reversal(self, route, route_demands , interdiction_probabilities, iter_num, tier_1_it, tier_2_it):
        best_route = route[:]
        best_distance = self.route_cost(route)
        it_number = iter_num+1
        
        osf_swl = self.optimization_solution_finder
        
        for i in range(1, len(route) - 2):
            for j in range(i + 1, len(route)):
                c1 = self.calculate_cumulative_probabilities_for_routes(best_route, interdiction_probabilities)
                new_route = route[:i] + list(reversed(route[i:j])) + route[j:]
                new_route_backup = new_route
                c2 = self.calculate_cumulative_probabilities_for_routes(new_route, interdiction_probabilities)
                current_distance = self.route_cost(new_route)

                old_route = best_route
                old_demands = route_demands

                new_route = self.check_and_transform_path(new_route)
                old_route = self.check_and_transform_path(old_route)
                old_demands = self.check_and_transform_demands(old_demands)
                  
                new_demands = self.reorder_demands_by_route(new_route, old_demands)
                new_demands_bu = new_demands  
                
                new_demands = self.check_and_transform_demands(new_demands)

            
                revised_demands_old_route = self.calculate_revised_demands_for_routes(old_route, old_demands, interdiction_probabilities)
                revised_demands_new_route = self.calculate_revised_demands_for_routes(new_route, new_demands, interdiction_probabilities)
                all_cumulative_probabilities_old_route = self.calculate_cumulative_probabilities_for_routes(old_route, interdiction_probabilities)
                all_cumulative_probabilities_new_route = self.calculate_cumulative_probabilities_for_routes(new_route, interdiction_probabilities)
                
                expected_demands_old_route = self.calculate_expected_demand_serviced(revised_demands_old_route, all_cumulative_probabilities_old_route)
                
                expected_demands_new_route = self.calculate_expected_demand_serviced(revised_demands_new_route, all_cumulative_probabilities_new_route)
                
                # Calculate the demands
                total_demand_by_route_check_old_route, number_of_routes_check_old_route, total_demand_check_old_route = self.calculate_demands_for_capacity_check(revised_demands_old_route)
                
                total_demand_by_route_check_new_route, number_of_routes_check_new_route, total_demand_check_new_route = self.calculate_demands_for_capacity_check(revised_demands_new_route)
                
                total_demand_check_new_route < total_demand_check_old_route
                route_demands = new_demands_bu
                
                if osf_swl == "tier":
                    if it_number <= tier_1_it:
                        if (round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_backup, route_demands):
                            
                            best_route = new_route_backup[:]
                                

                    elif it_number > tier_1_it and it_number <= tier_2_it:
                        if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_backup, route_demands):
                            
                            best_route = new_route_backup[:]
                                
                            
                    elif it_number > tier_2_it :
                        if current_distance < best_distance and self.is_route_valid(new_route_backup, route_demands):
                            best_distance = current_distance
                            best_route = new_route_backup[:]
                               

                elif osf_swl == "cost":
                    if current_distance < best_distance and self.is_route_valid(new_route_backup, route_demands):
                        best_distance = current_distance
                        best_route = new_route_backup[:]

                elif osf_swl == "inter_route":
                    if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_backup, route_demands):
                        
                        best_route = new_route_backup[:]
                
                elif osf_swl == "min_demand_route":
                    if (round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_backup, route_demands):
                        
                        best_route = new_route_backup[:]
                
                elif osf_swl == "random":
                    a = random.uniform(0, 1)
                    if a <= 0.33:
                        if (round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_backup, route_demands):
                            
                            best_route = new_route_backup[:]
                                

                    elif a > 0.33 and a <= 0.66:
                        if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_backup, route_demands):
                            
                            best_route = new_route_backup[:]
                                
                            
                    elif a > 0.66:
                        if current_distance < best_distance and self.is_route_valid(new_route_backup, route_demands):
                            best_distance = current_distance
                            best_route = new_route_backup[:]
                                

                elif osf_swl == "mixed":
                    b = random.uniform(0, 1)
                    if it_number <= tier_1_it and b <= 0.33:
                        if (round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_backup, route_demands):
                            
                            best_route = new_route_backup[:]
                                

                    elif it_number > tier_1_it and it_number <= tier_2_it and b <= 0.33:
                        if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_backup, route_demands):
                            
                            best_route = new_route_backup[:]
                                
                            
                    elif it_number > tier_2_it and b <= 0.33:
                        if current_distance < best_distance and self.is_route_valid(new_route_backup, route_demands):
                            best_distance = current_distance
                            best_route = new_route_backup[:]
                                

                    elif it_number <= tier_1_it and (b > 0.33 and b <= 0.66):
                        if (round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_backup, route_demands):
                            best_route = new_route_backup[:]
                                

                    elif it_number > tier_1_it and it_number <= tier_2_it and (b > 0.33 and b <= 0.66):
                        if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_backup, route_demands):
                            best_route = new_route_backup[:]
                                
                            
                    elif it_number > tier_2_it and (b > 0.33 and b <= 0.66):
                        if current_distance < best_distance and self.is_route_valid(new_route_backup, route_demands):
                            best_distance = current_distance
                            best_route = new_route_backup[:]
                                

                    elif it_number <= tier_1_it and b > 0.66:
                        if (round(total_demand_check_new_route,1) < round(total_demand_check_old_route,1)) and self.is_route_valid(new_route_backup, route_demands):
                            best_route = new_route_backup[:]
                                

                    elif it_number > tier_1_it and it_number <= tier_2_it and b > 0.66:
                        if (c2[0][0] < c1[0][0]) and self.is_route_valid(new_route_backup, route_demands):
                            best_route = new_route_backup[:]
                                
                            
                    elif it_number > tier_2_it and b > 0.66:
                        if current_distance < best_distance and self.is_route_valid(new_route_backup, route_demands):
                            best_distance = current_distance
                            best_route = new_route_backup[:]

        return best_route

    def local_search(self, routes, route_demands, interdiction_probabilities, iter_num, tier_1_it, tier_2_it):
        
        for idx, route in enumerate(routes):

            if self.local_search_option == '2-opt':
                routes[idx] = self.two_opt(route, route_demands[idx], interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
                
            elif self.local_search_option == '3-opt':
                routes[idx] = self.three_opt(route, route_demands[idx] , interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
            elif self.local_search_option == 'segment-reversal':
                routes[idx] = self.segment_reversal(route, route_demands[idx] , interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
            elif self.local_search_option == 'random':
                random_choice = random.choice(['2-opt', '3-opt', 'segment-reversal'])
                if random_choice == '2-opt':
                    # print("ENTERING 2-opt")
                    routes[idx] = self.two_opt(route, route_demands[idx] , interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
                    # print("FINISHING 2-opt")
                elif random_choice == '3-opt':
                    # print("ENTERING 3-opt")
                    routes[idx] = self.three_opt(route, route_demands[idx] , interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
                    # print("FINISHING 3-opt")
                elif random_choice == 'segment-reversal':
                    # print("ENTERING segment-reversal")
                    routes[idx] = self.segment_reversal(route, route_demands[idx] , interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
                    # print("FINISHING segment-reversal")
        # Apply swap-based local search after other local searches
        ordered_demands = self.reorder_demands_by_route(routes, route_demands)
        route_demands = ordered_demands
        
        # print("ENTERING swap")
        # print("Routes before swap: ", routes)
        # print("Demands before swap: ", route_demands)
        # print("Iter Num before swap: ", iter_num)
        # print("Tier 1 before swap: ", tier_1_it)
        # print("Tier 2 before swap: ", tier_2_it)
        self.swap_local_search(routes, route_demands, interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
        # print("FINISHING swap")
        return routes, route_demands
        
    # Function to reorder demands to match the route order
    def reorder_demands_by_route(self, routes, demands):
        ordered_demands = []
        for route, demand in zip(routes, demands):
            ordered_route_demands = {}
            # Skip the first and last element assuming they're the depot (0)
            for node in route[1:-1]:
                if node in demand:
                    ordered_route_demands[node] = demand[node]
            ordered_demands.append(ordered_route_demands)
        return ordered_demands

    def route_cost(self, route):
        total_distance = sum(self.distance_matrix[route[i], route[i+1]] for i in range(len(route)-1))
        # We'll apply the penalty in the grasp function when summing up all routes cost
        return total_distance
    
    # Function to check and transform path
    def check_and_transform_path(self, path):
        if isinstance(path, list) and all(isinstance(item, list) for item in path):
            # path is already a list of lists
            return path
        elif isinstance(path, list):
            # path is a single list
            return [path]
        else:
            raise ValueError("path is not in the expected format")

    # Function to check and transform demands_list
    def check_and_transform_demands(self, demands_list):
        if isinstance(demands_list, list) and all(isinstance(item, dict) for item in demands_list):
            # demands_list is already a list of dictionaries
            return demands_list
        elif isinstance(demands_list, dict):
            # demands_list is a single dictionary
            return [demands_list]
        else:
            raise ValueError("demands_list is not in the expected format")
    
    
    def check_sdvrp_feasibility(self, routes, serviced_demands, demands, num_vehicles, vehicle_capacity):
        num_customers = len(demands)
        total_demands_satisfied = {i+1: 0 for i in range(num_customers)}
        customers_serviced = set()

        # Condition 2: Check if number of routes matches number of vehicles
        if len(routes) != num_vehicles:
            return False, "The number of routes does not match the number of vehicles."

        for vehicle, route_demands in enumerate(serviced_demands):
            route_load = 0  # Total load on the current route

            for customer, amount in route_demands.items():
                route_load += amount
                total_demands_satisfied[customer] += amount
                customers_serviced.add(customer)  # Track which customers are serviced

                # Condition 4: Check if any delivery to a customer exceeds vehicle capacity
                if amount > vehicle_capacity:
                    return False, f"Delivery to customer {customer} exceeds vehicle capacity on route {vehicle+1}."

            # Condition 1: Check if total demand on route exceeds vehicle capacity
            if route_load > vehicle_capacity:
                return False, f"Total demand on route {vehicle+1} exceeds vehicle capacity."

        # Condition 3: Check if all customers are serviced
        if len(customers_serviced) != num_customers:
            return False, "Not all customers are serviced."

        # Condition 5: Check if all demands are met or exceeded
        for customer, demand in enumerate(demands, start=1):
            if total_demands_satisfied[customer] < demand:
                return False, f"Customer {customer}'s total received demand {total_demands_satisfied[customer]} is less than required {demand}."

        return True, "All conditions satisfied."
    
    def check_vrpi_feasibility(self, routes, revised_demands, expected_demands, demands, num_vehicles, vehicle_capacity):
        num_customers = len(demands)
        total_demands_satisfied = {i+1: 0 for i in range(num_customers)}
        customers_serviced = set()

        # Condition 2: Check if number of routes matches number of vehicles
        if len(routes) != num_vehicles:
            return False, "The number of routes does not match the number of vehicles."

        for vehicle, route_demands in enumerate(revised_demands):
            route_load = 0  # Total load on the current route

            for customer, amount in route_demands.items():
                route_load += amount
                total_demands_satisfied[customer] += amount
                customers_serviced.add(customer)  # Track which customers are serviced

                # Condition 4: Check if any delivery to a customer exceeds vehicle capacity
                if amount > vehicle_capacity:
                    return False, f"Delivery to customer {customer} exceeds vehicle capacity on route {vehicle+1}."

            # Condition 1: Check if total demand on route exceeds vehicle capacity
            if route_load > vehicle_capacity:
                return False, f"Total demand on route {vehicle+1} exceeds vehicle capacity."

        # Condition 3: Check if all customers are serviced
        if len(customers_serviced) != num_customers:
            return False, "Not all customers are serviced."

        # Condition 5: Check if all demands are met or exceeded
        # Iterate over the demands list with index (1-based)
        for index, demand in enumerate(demands, start=1):
            expected_demand = expected_demands.get(index, 0)  # Get expected demand or 0 if not found
            # Check if the expected demand meets or exceeds the specified demand
            if expected_demand < demand:
                return False, f"Expected demand for customer {index} is {expected_demand}, which is less than required {demand}."

        return True, "All conditions satisfied."

    def calculate_increments(self, total_demand, increment_threshold):
        increment = total_demand / increment_threshold
        return [i * increment for i in range(increment_threshold + 1)]

    def generate_combinations(self, routes, demands, increment_threshold, interdiction_probabilities, input_demands):
        # Identify customers with split deliveries and calculate their total demands
        split_customers = {}
        total_demands = {}

        
        for route_idx, route in enumerate(routes):
            for customer in route[1:-1]:  # Exclude the depot (0)
                #print(route[1:-1])
                if customer not in total_demands:
                    total_demands[customer] = 0
                total_demands[customer] += demands[route_idx].get(customer, 0)

        for customer, total_demand in total_demands.items():
            route_indices = [route_idx for route_idx, route in enumerate(routes) if customer in demands[route_idx]]
            if len(route_indices) > 1:
                split_customers[customer] = (total_demand, route_indices)
        
        # Generate demand increments for each split customer
        customer_combinations = {}
        for customer, (total_demand, route_indices) in split_customers.items():
            increments = self.calculate_increments(total_demand, increment_threshold)
            customer_combinations[customer] = list(itertools.product(increments, repeat=len(route_indices)))
            # Filter out combinations that don't sum up to the total demand
            customer_combinations[customer] = [comb for comb in customer_combinations[customer] if abs(sum(comb) - total_demand) <= 1e-6]
        
        # Generate all possible combinations
        all_combinations = list(itertools.product(*customer_combinations.values()))
        
        # Initialize error count
        error_count = 0

        
        
        # Process each combination
        for combination_set in all_combinations:
            combination_demands = [{k: v for k, v in route.items()} for route in demands]
            
            for idx, (customer, (total_demand, route_indices)) in enumerate(split_customers.items()):
                combination = combination_set[idx]
                for route_idx, demand in zip(route_indices, combination):
                    combination_demands[route_idx][customer] = demand
            
            all_cumulative_probabilities = self.calculate_cumulative_probabilities_for_routes(routes, interdiction_probabilities)
            revised_demands = self.calculate_revised_demands_for_routes(routes, combination_demands, interdiction_probabilities)
            expected_demands = self.calculate_expected_demand_serviced(revised_demands, all_cumulative_probabilities)
            result_f, message = self.check_vrpi_feasibility(routes, revised_demands, expected_demands, input_demands, num_vehicles, vehicle_capacity)
            total_serviced_demands = {}
            for customer, total_demand in total_demands.items():
                serviced_demand = sum(combination_demands[route_idx].get(customer, 0) for route_idx in range(len(routes)))
                total_serviced_demands[customer] = serviced_demand
                log = True
                if abs(serviced_demand - total_demand) > 1e-6:
                    print(f"Error: Serviced demand for customer {customer} does not match total demand.")
                    print(f"Expected: {total_demand}, Got: {serviced_demand}")
                    error_count += 1
                    log = False
                    
                if any(demand < 0 for demand in (combination_demands[route_idx].get(customer, 0) for route_idx in range(len(routes)))):
                    print(f"Error: Negative demand for customer {customer} in one of the routes.")
                    error_count += 1
                    log = False
            if result_f is True and log is True:
                return "we have a candidate", routes, revised_demands
            
            
        
        print(f"Total errors: {error_count}")
        return "we did not find a candidate", routes, revised_demands    
        
        
    
    def calculate_inverse_cumulative_probability(self, path, interdiction_probabilities):
        cumulative_probabilities = {path[0]: 1}  # Start with depot having a cumulative probability of 1
        for i in range(1, len(path)):
            cumulative_probabilities[path[i]] = cumulative_probabilities[path[i-1]] * interdiction_probabilities[path[i-1]][path[i]]
        inverse_cumulative_probabilities = {node: 1/prob for node, prob in cumulative_probabilities.items() if prob != 0}
        return inverse_cumulative_probabilities

    def calculate_revised_demands_for_routes(self, routes, demands, interdiction_probabilities):
        all_revised_demands = []
        for route, demand_serviced in zip(routes, demands):
            # Calculate inverse cumulative probabilities for the route
            inverse_cumulative_probabilities = self.calculate_inverse_cumulative_probability(route, interdiction_probabilities)
            
            # Calculate revised demands
            revised_demands = {}
            for node, demand in demand_serviced.items():
                if node in inverse_cumulative_probabilities:
                    revised_demands[node] = demand * inverse_cumulative_probabilities[node]
            
            all_revised_demands.append(revised_demands)
        
        return all_revised_demands
    
    def calculate_cumulative_probabilities_for_routes(self, routes, interdiction_probabilities):
        # Ensure that 'routes' is always a list of lists
        if not isinstance(routes[0], list):
            routes = [routes]  # Wrap it in another list if it's not already a list of lists
        
        all_cumulative_probabilities = []
        for path in routes:
            cumulative_probabilities = {path[0]: 1}  # Start with depot having a cumulative probability of 1
            for i in range(1, len(path)):
                # Compute cumulative probability for the current node on the path
                cumulative_probabilities[path[i]] = cumulative_probabilities[path[i-1]] * interdiction_probabilities[path[i-1]][path[i]]
            # Append the calculated cumulative probabilities for this route to the list
            all_cumulative_probabilities.append(cumulative_probabilities)
        return all_cumulative_probabilities
    
    def is_matrix_uniform(self, interdiction_probabilities):
        """Check if the interdiction matrix has all elements as 1."""
        for row in interdiction_probabilities:
            for prob in row:
                if prob != 1:
                    return False
        return True

    
    def calculate_expected_demand_serviced(self, revised_demands, cumulative_probabilities):
    # Dictionary to store the expected demand serviced by each customer
        expected_demand_serviced = {}
        
        # Process each route's data
        for revised, probabilities in zip(revised_demands, cumulative_probabilities):
            for node in revised:
                if node in probabilities:
                    # Calculate expected demand for the node
                    expected_demand = revised[node] * probabilities[node]
                    
                    # Add to existing demand if node appears in multiple routes
                    if node in expected_demand_serviced:
                        expected_demand_serviced[node] += expected_demand
                    else:
                        expected_demand_serviced[node] = expected_demand
        #expected_demand_serviced = {key: value + 1e-10 for key, value in expected_demand_serviced.items()}
        expected_demand_serviced = {key: round(value,2) for key, value in expected_demand_serviced.items()}
        return expected_demand_serviced
    

    def is_square_and_dimensions(self, matrix):
        # Determine the number of rows
        num_rows = len(matrix)
        
        # Determine if all rows have the same length as the number of rows (i.e., it's a square matrix)
        is_square = all(len(row) == num_rows for row in matrix)
        
        # Dimensions are given by the number of rows and the number of columns (assuming it's a square)
        dimensions = (num_rows, num_rows if is_square else None)  # Set columns to None if it's not a square
        
        return is_square, dimensions
    
    def generate_intr_matrix(self, n, a, b):
        # Generate the upper triangle of the matrix with random numbers between min and max
        upper_triangle = np.triu(np.random.uniform(a, b, (n, n)), 1)

        # Create the symmetric matrix by adding the transpose of the upper triangle to itself and the diagonal as 1
        matrix = upper_triangle + upper_triangle.T + np.eye(n)

        # Convert numpy array to a list of lists
        interdiction_probabilities = matrix.tolist()
        return interdiction_probabilities
    
    def is_all_ones(self, matrix, tolerance=1e-9):
        for row in matrix:
            for value in row:
                if abs(value - 1.0) > tolerance:
                    return False
        return True

    def grasp(self):
        start_time = time.time()
        best_routes = None
        best_route_demands = None
        best_cost = float('inf')
        no_improvement_counter = 0
        potential_solution_counter = 0
        sl = 'No'
        if self.int_prob is not None:
            is_square, dimensions = self.is_square_and_dimensions(self.int_prob)
            if dimensions[0] == dimensions [1] and dimensions[0] == num_customers+1:
                interdiction_probabilities = self.int_prob
                print("Interdiction Probabilities: ",interdiction_probabilities)
            else:
                print("The manual interdiction probabilities matrix does not have the right dimensions for this problem")
                print("It was replaced by a matrix full of 1's") 
                interdiction_probabilities = self.generate_intr_matrix(num_customers+1, 1.0, 1.0)
                print(interdiction_probabilities)
        else:
            interdiction_probabilities = self.generate_intr_matrix(num_customers+1, self.min_prob_matrix, self.max_prob_matrix)
            print(interdiction_probabilities)

        tier_1_it = round(self.iterations/3,0)
        tier_2_it = round(self.iterations*2/3,0)
        

        for iter_num in range(self.iterations):
            if iter_num != 0 and iter_num >= self.reintegrate_it:
                # Reintegrate the saved initial solution
                routes, route_demands, remaining_demands = self.construct_solution(reintegrate=True)
                
            else:
                # Normal solution generation
                routes, route_demands, remaining_demands = self.construct_solution()
                
            # print("ENTERING LOCAL SEARCH")
            routes, route_demands = self.local_search(routes, route_demands, interdiction_probabilities, iter_num, tier_1_it, tier_2_it)
            
            revised_demands = self.calculate_revised_demands_for_routes(routes, route_demands, interdiction_probabilities)
            all_cumulative_probabilities = self.calculate_cumulative_probabilities_for_routes(routes, interdiction_probabilities)
            
            input_demands = demands
            
            

            result_matrix = self.is_all_ones(interdiction_probabilities)
            
            if result_matrix is True:
                result_f, message = self.check_sdvrp_feasibility(routes, route_demands, demands, num_vehicles, vehicle_capacity)
                print("This is a candidate for SDVRP")
            elif result_matrix is False:
                expected_demands = self.calculate_expected_demand_serviced(revised_demands, all_cumulative_probabilities)
                result_f, message = self.check_vrpi_feasibility(routes, revised_demands, expected_demands, demands, num_vehicles, vehicle_capacity)
                print("This is a candidate for VRPI")

            

            if result_f is False:
                message2, routes, route_demands = self.generate_combinations(routes, route_demands, self.threshold_increment, interdiction_probabilities, input_demands)
                print("Message 2: ", message2)
                # if message == "we have a candidate":
                if message2 == "we have a candidate":
                    result_f = True
                    revised_demands = route_demands
                else:
                    result_f = False
                    revised_demands = revised_demands
            else:
                print("No need to generate combinations")
                
            

            split_deliveries = self.count_split_deliveries(routes, revised_demands)
            cost = sum(self.route_cost(route) for route in routes) + split_deliveries * self.split_penalty
            print("Iteration: ", iter_num, ". Cost: ", cost)
            #print((cost/known-1), '-->', thres)
            div = round(self.iterations/12,0)
            if (iter_num+1) % div == 0:
                print('Iteration # ', iter_num+1, ' out of ', self.iterations,
                  ' (', round((iter_num+1)/self.iterations,2)*100, '%)')
                print('Elapsed time: ', round(time.time() - start_time,0), ' seconds.')

            if ((cost < known) and (result_f is True or message2 == "we have a candidate")) or (((cost/known-1) <= thres) and (result_f is True or message2 == "we have a candidate")):
            #if (cost < known) or (((cost/known-1) <= thres)):
                #print('True: ', (cost/known-1) )
                improvement_time = time.time() - start_time
                trigger = cost/known-1
                best_cost = cost
                best_routes = routes
                best_route_demands = revised_demands
                potential_solution_counter += 1
                if cost < known and result_f is True:
                    print("Early finish, best cost found was lower than known value. Best cost: ", best_cost, ", and known cost is: ", known)
                else:
                    print("Early finish, difference between known value and best cost was: ", str(trigger*100)+'%',
                 ', which is less than the threshold: ', str(thres*100)+'%')
                break
            #print("Cost is ",cost, ", and Best Cost is ", best_cost)
            if cost < best_cost and (result_f is True or message2 == "we have a candidate"):
                improvement_time = time.time() - start_time
                best_cost = cost
                best_routes = routes
                best_route_demands = revised_demands
                no_improvement_counter = 0
                potential_solution_counter += 1
            else:
                no_improvement_counter += 1

            if no_improvement_counter == int(self.iterations * pert_ct):
                self.perturb(routes)

            if no_improvement_counter >= self.stagnation_limit:
                print("Early termination due to stagnation.")
                sl = 'Yes'
                break

        finish_time = time.time() - start_time

        
        if potential_solution_counter >0:
            improvement_time = improvement_time
        else:
            improvement_time = 1e20
            best_routes = "NA"
            best_cost = "NA"
            best_route_demands = "NA"
            remaining_demands = "NA"

    

        return best_routes, best_cost, best_route_demands, remaining_demands, improvement_time, finish_time, sl

    def perturb(self, routes):
        k = int(self.number_customers * self.perturb_percentage)
        perturbed_customers = random.sample(range(1, self.number_customers + 1), k)
        #print('pert customers',perturbed_customers)

        for customer in perturbed_customers:
            for route in routes:
                if customer in route:
                    route.remove(customer)

        for customer in perturbed_customers:
            insertion_route = random.choice(routes)
            insertion_position = random.randint(1, len(insertion_route) - 1)
            #print('inst route', insertion_route)
            #print('inst pos', insertion_position)
            insertion_route.insert(insertion_position, customer)


    def print_results(self, best_routes, best_cost, best_route_demands, remaining_demands):
        for i, route in enumerate(best_routes):
            print(f"Route {i + 1}: {' -> '.join(map(str, route))}")
            print(f"Cost: {self.route_cost(route):.2f}")

        print("Total cost:", best_cost)

        print("\nServiced demands for each route:")
        #print("Best routes demands: ",best_route_demands)
        for i, demands_dict in enumerate(best_route_demands):
            print(f"Route {i + 1} Total Demand Serviced: {sum(demands_dict.values())}")
            for customer in best_routes[i][1:-1]:
                if customer in demands_dict:
                    print(f"   Customer {customer}: {demands_dict[customer]}")

        print("\nCustomers with split delivery:")

        split_deliveries = {}
        for i, demands_dict in enumerate(best_route_demands):
            for customer, demand in demands_dict.items():
                if customer not in split_deliveries:
                    split_deliveries[customer] = []
                split_deliveries[customer].append((best_routes[i], demand))

        for customer, deliveries in split_deliveries.items():
            if len(deliveries) > 1:
                print(f"  Customer {customer}:")
                for delivery in deliveries:
                    route_str = " -> ".join(map(str, delivery[0]))
                    print(f"    Demand serviced by route ({route_str}): {delivery[1]}")

        # Identify and print customers who were not served (either partially or totally)
        print("\nCustomers not served or partially served:")
        for customer_id, rem_demand in enumerate(remaining_demands):
            if rem_demand > 0 and customer_id != 0:  # Exclude the depot with customer_id = 0
                print(f"  Customer {customer_id}: Unserved Demand {rem_demand}")

        # Count clients with split delivery
        split_clients = [cust for cust, deliveries in split_deliveries.items() if len(deliveries) > 1]
        num_split_clients = len(split_clients)

        # Count total clients and clients without split delivery
        total_clients = self.number_customers
        num_non_split_clients = total_clients - num_split_clients

        return num_split_clients, num_non_split_clients, total_clients

    def run(self):
        best_routes, best_cost, best_route_demands, remaining_demands, improvement_time, finish_time, sl = self.grasp()
        if best_routes == "NA":
            num_split_clients = "NA"
            num_non_split_clients = "NA"
            total_clients = "NA"
            solution = "We did not find a feasible solution"
        else:
            num_split_clients, num_non_split_clients, total_clients = self.print_results(best_routes, best_cost, best_route_demands, remaining_demands)
            solution = "We found a feasible solution"

        print(f"\nNumber of clients with split deliveries: {num_split_clients}")
        print(f"Number of clients without split deliveries: {num_non_split_clients}")
        print(f"Total number of clients: {total_clients}")

        return best_cost, num_split_clients, num_non_split_clients, total_clients, improvement_time, finish_time, sl, solution  # Ensure you return the best_cost along with the counts
    
    
```
